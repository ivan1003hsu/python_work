```python
import tkinter as tk
import tensorflow as tf
from tensorflow.keras.layers import Dense, LSTM, GRU, Conv2D, MaxPooling2D, Input
from tensorflow.keras.models import Sequential
```

```python
import tkinter as tk
```

```python
window = tk.Tk()
window.title('Sequential')
window.geometry('600x800')

frm = tk.Frame(window)
frm.pack()

frm1 = tk.Frame(frm)
frm1.pack(side='top')
frm1_1 = tk.Frame(frm1)
frm1_2 = tk.Frame(frm1)
frm1_3 = tk.Frame(frm1)
frm1_1.pack(side='left')
frm1_2.pack(side='left')
frm1_3.pack(side='left')

frm2 = tk.Frame(frm)
frm2.pack(side='top')


l1 = tk.Label(frm1_1, text='current choose:', width=10, height=2).pack(side='left')

var1 = tk.StringVar()

l2 = tk.Label(frm1_2, textvariable=var1, bg='yellow', font=('Arial',12), width=15, height=2)
l2.pack(side='left')

l3 = tk.Label(frm1_2, text='untis:')
l3.pack(side='left')
 
units = tk.Entry(frm1_2, bg='yellow', width=15)
units.pack(side='left')

l4 = tk.Label(frm1_3, text='fliter size:')

fliter_s = tk.Entry(frm1_3, bg='yellow', width=15)


def show_units():
    l3.config(text='unitls:')
    l3.pack(side='left')
    units.pack(side='left')
    l4.pack_forget()
    fliter_s.pack_forget()
def show_f_n_s():
    l3.config(text='fliter number:')
    l3.pack(side='left')
    units.pack(side='left')
    l4.config(text='fliter size:(a,b)')
    l4.pack(side='left')
    fliter_s.pack(side='left')
def show_p_s():
    l3.pack_forget()
    units.pack_forget()
    l4.config(text='pooling size:(a,b)')
    l4.pack(side='left')
    fliter_s.pack(side='left')
def show_nothing():
    l3.pack_forget()
    units.pack_forget()
    l4.pack_forget()
    fliter_s.pack_forget()
    
rb1 = tk.Radiobutton(frm2, text='Dense', variable=var1, value='Dense'
                    ,command=show_units)
rb1.pack(side='left')
rb2 = tk.Radiobutton(frm2, text='LSTM', variable=var1, value='LSTM'
                    ,command=show_units)
rb2.pack(side='left')
rb3 = tk.Radiobutton(frm2, text='GRU', variable=var1, value='GRU'
                    ,command=show_units)
rb3.pack(side='left')
rb4 = tk.Radiobutton(frm2, text='Conv2D', variable=var1, value='Conv2D'
                    ,command=show_f_n_s)
rb4.pack(side='left')
rb5 = tk.Radiobutton(frm2, text='Maxpooling2D', variable=var1, value='Maxpooling2D'
                    ,command=show_p_s)
rb5.pack(side='left')
rb6 = tk.Radiobutton(frm2, text='Flatten', variable=var1, value='Flatten'
                    ,command=show_nothing)
rb6.pack(side='left')

frm3 = tk.Frame(frm)
frm3.pack(side='top')
frm3_1 = tk.Frame(frm3)
frm3_2 = tk.Frame(frm3)
frm3_1.pack(side='left')
frm3_2.pack(side='left')

def lb_add_layer():
    latype=var1.get()
    if latype=='Dense' or latype=='LSTM' or latype=='GRU':
        lb.insert('end', var1.get() + ' ' + units.get() + ' ')
        
    elif latype=='Conv2D':
        lb.insert('end', var1.get() + ' ' + units.get() + ' ' + fliter_s.get())
        
    elif latype=='Maxpooling2D':
        lb.insert('end', var1.get() + ' '               + ' ' + fliter_s.get())
        
    else:
        lb.insert('end', var1.get() + ' '               + ' ')
        
def lb_del_layer():
    lb.delete(lb.curselection()[0])
add_lay = tk.Button(frm3_1,text ='add layer', command=lb_add_layer).pack()
del_lay = tk.Button(frm3_2,text ='delete layer', command=lb_del_layer).pack()



frm4 = tk.Frame(frm)
frm4.pack(side='top')
lb = tk.Listbox(frm4, width=20, height=22)
lb.pack()

frm5 = tk.Frame(frm)
frm5.pack(side='top')
frm5_1 = tk.Frame(frm5)
frm5_2 = tk.Frame(frm5)
frm5_1.pack(side='left')
frm5_2.pack(side='left')

l5 = tk.Label(frm5_1, text='Input dim:').pack()
l6 = tk.Label(frm5_1, text='Model name:').pack()
input_shape= tk.Entry(frm5_2, bg='yellow')
input_shape.pack()
model_name= tk.Entry(frm5_2, bg='yellow')
model_name.pack()

frm6 = tk.Frame(frm)
frm6.pack(side='top')
frm6_1 = tk.Frame(frm6)
frm6_2 = tk.Frame(frm6)
frm6_1.pack(side='left')
frm6_2.pack(side='left')

def add_layer(layer_type,unit,fliter):
    global model
    if layer_type =='Dense':
        model.add(Dense(unit, activation='relu'))
    elif layer_type=='LSTM':
        model.add(LSTM(unit))
    elif layer_type=='GRU':
        model.add(GRU(unit))
    elif layer_type=='Conv2D':
        model.add(Conv2D(unit, fliter, padding='same', activation='relu') )
    elif layer_type=='Maxpooling2D':
        model.add(MaxPooling2D(pool_size=fliter))
    else:
        model.add(flatten())

def build_model():
    global model
    model = Sequential()
    model.add(Input(shape=( int(input_shape.get())), ) )
    for i in range(lb.size()):
        la = lb.get(i)
        layer_type, unit, fliter = la.split(' ')
        try:
            unit = int(unit)
        except:
            pass
        try:
            fliter = (int(fliter[0], int(fliter[1])))
        except:
            pass
        add_layer(layer_type,unit,fliter)
    model.compile(loss='mse', optimizer='adam', metrics=['accuracy'])
    model.summary()
    
b_bm = tk.Button(frm6_1, text='Build Model', command=build_model).pack()

def save_model():
    model.save( model_name.get() +'.h5')
b_sm = tk.Button(frm6_2, text='Save Model', command=save_model).pack()
    
window.mainloop()
```
