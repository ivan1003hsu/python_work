```python
from urllib.request import urlopen
```

```python
html = urlopen('https://morvanzhou.github.io/static/scraping/basic-structure.html').read().decode('utf-8')
```

```python
print(html)
```

```python
from bs4 import BeautifulSoup
import time
```

```python
soup = BeautifulSoup(html, features='lxml')
print(soup.h1,end='')
time.sleep(2)
print('\r                 ',end='')
print('\r',soup.p)
```

```python
all_href = soup.find_all('a')
print(all_href)
for l in all_href:
    print(l['href'])
```

```python
school = urlopen("https://www.nccu.edu.tw/app/home.php").read().decode('utf-8')
sch_soup = BeautifulSoup(school, features='lxml')
print(sch_soup)
```

```python
all_href_in_school = sch_soup.find_all('a')
for l in all_href_in_school:
    print(l['href'])
```
